```python
df["añodevencimiento"] = df["fechavencimiento"].astype(str).str[-4:].astype(int)

minimo_añodevencimiento = np.min(df["añodevencimiento"])
maximo_añodevencimiento = np.max(df["añodevencimiento"])
print(f"Año más próximo de vencimiento: {minimo_añodevencimiento}")
print(f"Año más lejano de vencimiento: {maximo_añodevencimiento}")
```

```python
promedio_cantidad = np.mean(df["cantidad"])
print(f"Promedio de cantidad presentada: {promedio_cantidad}")
```

```python
mediana_cantidad = np.median(df["cantidad"])
print(f"Mediana de cantidad presentada: {mediana_cantidad}")
```
