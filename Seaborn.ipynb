```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\gusta\OneDrive\Documentos\Proyecto_Ciencia_de_Datos_Grupo_5\Trabajo_Acompañamiento_Grupo5_dF.csv")
df
```

```python
sns.set(style="dark", palette= "bright")
plt.figure(figsize=(10, 6))

sns.countplot(x= "estadodeformafarmaceutica", data=df)
plt.title("Cantidad de productos POS según su estado de forma farmacéutica")
plt.xlabel("estado de forma farmaceutica")
plt.ylabel("cantidad")
```

```python
df["añodevencimiento"] = df["fechavencimiento"].astype(str).str[-4:].astype(int)
df
```

```python
sns.set(palette="bright")

sns.boxplot(x="añodevencimiento", y="estadodeformafarmaceutica", data=df)
plt.title("Distribución de productos según su estado de forma farmacéutica por año de vencimiento")
plt.xlabel("Año de Vencimiento")
plt.ylabel("Estado de Forma Farmacéutica")

plt.xticks(rotation=45)

plt.show()
```

```python
conteo=df["concentracion"].value_counts()
conteo
```

```python
sns.set(style="darkgrid", palette="bright")

plt.figure(figsize=(12, 8))

sns.countplot(x="concentracion", hue="estadodeformafarmaceutica", data=df)
plt.title("Relación de la cantidad de productos según su estado de forma farmacéutica y la concentración")
plt.xlabel("Concentración")
plt.ylabel("Cantidad")

plt.xticks(rotation=45)
plt.legend(title="Estado de forma farmacéutica")

plt.show()
```
