```python
from unidecode import unidecode

def eliminar_acentos(columna):
    return columna.apply(lambda x: unidecode(x) if isinstance(x, str) else x)

df = df.apply(eliminar_acentos)

df
```

```python
df = df.astype(str)
```

```python
columnas_con_puntos = ["cantidad","nombrerol"]

def quitar_puntos(columna):
    if columna.name not in columnas_con_puntos:
        return columna.str.replace('.', '')
    else:
        return columna
    
df = df.apply(quitar_puntos)
```

```python
recuento_valores = df["formafarmaceutica"].value_counts()
recuento_valores
```

```python
def clasificar_medicamento(formafarmaceutica):
    if formafarmaceutica in ["TABLETA RECUBIERTA", "TABLETA", "CAPSULA DURA", "TABLETA CUBIERTA CON PELICULA", "POLVO LIOFILIZADO PARA RECONSTITUIR A SOLUCION INYECTABLE", "POLVO ESTERIL PARA RECONSTITUIR A SOLUCION INYECTABLE", "POLVO PARA RECONSTITUIR A SOLUCION ORAL", "TABLETA DE LIBERACION PROLONGADA", "POLVO PARA RECONSTITUIR A SUSPENSION ORAL", "GRANULOS", "CAPSULA DE LIBERACION PROLONGADA", "TABLETA CUBIERTA (GRAGEA)", "OVULO", "TABLETA DE LIBERACION RETARDADA", "POLVO ESTERIL PARA RECONSTITUIR A SUSPENSIONINYECTABLE", "TABLETA DISPERSABLE", "GRANULOS EFERVESCENTES", "TABLETA CON CUBIERTA ENTERICA CON PELICULA", "COMPRIMIDO", "CAPSULA CON CUBIERTA ENTERICA", "POLVO PARA INHALACION", "TABLETA CON RECUBRIMIENTO ENTERICO", "TABLETA EFERVESCENTE", "TABLETA ORODISPERSABLE", "TABLETA BUCAL", "POLVO PARA SOLUCION INYECTABLE", "POLVO LIOFILIZADO", "TRANSDERMICOS", "TABLETA CON CUBIERTA ENTERICA(GRAGEA)", "POLVO", "TABLETA DE LIBERACION MODIFICADA", "SUPOSITORIO", "TABLETA SUBLINGUAL", "SISTEMAS DE LIBERACION", "POLVO MEDICADO PARA USO TOPICO", "IMPLANTES Y SISTEMAS INTRAUTERINOS E INTRAOCULARES", "SISTEMA TRANSDERMICO DE LIBERACION PROLONGADA (PARCHE)", "CAPSULA VAGINAL", "IMPLANTE", "LIOFILIZADO ORAL DE DISPERSION RAPIDA", "CAPSULAS DURAS Y BLANDAS" ]:
        return "sólido"
    elif formafarmaceutica in ["CAPSULA BLANDA", "TABLETA MASTICABLE", "CREMA TOPICA", "GEL TOPICO", "UNGUENTO TOPICO", "CREMA VAGINAL", "LOCION", "EMULSION", "AEROSOLES", "UNGUENTO OFTALMICO", "POMADA", "EMULSION (CHAMPOO)", "CREMAS GELES UNGUENTOS Y PASTAS", "PASTA", "EMULSION ORAL", "GEL ORAL", "CAPSULA BLANDA DE GELATINA CON CUBIERTA ENTERICA", "CAPSULA BLANDA TWIST OFF", "GEL ESTERIL INTRAOCULAR", "GEL VAGINAL", "JALEA" ]:
        return "semisólido"
    elif formafarmaceutica in ["SOLUCION INYECTABLE", "SOLUCION ORAL", "EMULSION INYECTABLE", "SOLUCION OFTALMICA", "JARABE", "SOLUCION TOPICA", "SOLUCION ESTERIL PARA DIALISIS PERITONEAL", "SOLUCION CONCENTRADA PARA INFUSION", "SOLUCION ESTERIL PARA IRRIGACION", "INYECTABLES", "SOLUCION BUCAL", "SOLUCIONNASAL", "SOLUCION OTICA", "SOLUCION PARA INHALACION", "SOLUCION PARA INFUSION", "LIQUIDO ESTERIL PARA INYECCION (AGUA)", "SOLUCION BUCOFARINGUEA", "SOLUCION PARA ADMINISTRAR POR VIA RECTAL", "SOLUCION PARA NEBULIZACION", "ELIXIR", "SOLUCION SUBLINGUAL"]:
        return "líquido"
    elif formafarmaceutica in ["GAS"]:
        return "gas"
    elif formafarmaceutica in ["SUSPENSION ORAL", "SUSPENSION INYECTABLE", "SUSPENSION OFTALMICA", "SUSPENSIONES", "SUSPENSION PARA INHALACION", "SUSPENSION NASAL", "SUSPENSION PARA NEBULIZACION", "SUSPENSION OTICA", "SUSPENSION PARA ADMINISTRAR POR VIA RECTAL", "SUSPENSION TOPICA"]:
        return "sistema heterodisperso"
    elif formafarmaceutica in ["JABON LIQUIDO", "ESPONJA MEDICADA", "SIN DATO"]:
        return "otras/especiales"
    
df["estadodeformafarmaceutica"] = df["formafarmaceutica"].apply(clasificar_medicamento)
df
```
